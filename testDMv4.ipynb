En el siguiente ejemplo, se crea un archivo "csv" que contiene un array declarado en la variable "a". Para el siguiente ejemplo, se realiza la Suma de un DataFrame, compuesto por tres arreglos. Para ello, se declara la variable que contiene los arrays y se lo declara en otra variable, como un DataFrame, para que, con el comando "SUM", realice la suma de los números que componen el array, de manera que el resultado es un número escalar.  

```python
frame_data = {'A': [23, 12, 12], 'B': [18, 18, 22], 'C': [13, 112, 13]}
df = pd.DataFrame(frame_data)
df['A'].sum()
```

## Librería Seaborn

```python
import seaborn as sns
titanic_dataset = sns.load_dataset("titanic")
print(titanic_dataset.head())
```

```python
sns.set_style('darkgrid')
sns.violinplot(x = "sex", y = "survived", data = titanic_dataset)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.boxplot(x = "sex", y = "age", data = titanic_dataset)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.catplot(x='survived', y='age', hue='sex', kind="swarm", data= titanic_dataset)
plt.show()
```
